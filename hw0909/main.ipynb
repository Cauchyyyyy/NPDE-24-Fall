```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
v = lambda x: -1/2*(x-np.pi)

# v_N is the first N sum of the Fourier series of v
def v_N(x, N):
    return sum([np.sin(n*x)/n for n in range(1, N+1)])

# v_tildeN is the first N sum of the Fourier series of v with correction
def v_tildeN(x, N):
    return sum([np.sin(n*x)*np.sin(n*np.pi/N)/n**2 for n in range(1, N+1)])*N/np.pi
```

```python
x = np.linspace(0, 2 * np.pi, 20)
N = 10
plt.plot(x, v(x), label="v(x)")
plt.plot(x, v_N(x, N), label="v_N(x)")
plt.plot(x, v_tildeN(x, N), label="v_tildeN(x)")
plt.title("Fourier Series Approximation of v(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()

plt.plot(x, v(x) - v_N(x, N), label="Error v - v_N")
plt.plot(x, v(x) - v_tildeN(x, N), label="Error v - v_tildeN")
plt.title("Error in Fourier Series Approximation")
plt.grid()
plt.legend()
plt.show()
```
